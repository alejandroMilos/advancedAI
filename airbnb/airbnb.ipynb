```python
# Importacion de libreria polars y lectura de archivo CSV
import polars as pl
df = pl.read_csv("./listings.csv.gz")

# Mostrar la tabla en pantalla
df
```

```python
# Informacion sobre el csv
num_filas = df.height
num_cols = df.width
nombres_columnas = df.columns

print(f"Número de filas: {num_filas}")
print(f"Número de columnas: {num_cols}")
```

```python
# Filtrado y limpieza de la columna de precios

df = df.drop_nulls("price")
df = df.with_columns(pl.col("price").cast(pl.Utf8).str.replace_all(r"[\$,]", "").cast(pl.Float64))
print("Columna de precios limpia:")
print(df.select("price"))
```

```python
# Encontrar el room_type más común
rt_count = df["room_type"].value_counts().sort("count", descending=True)

rt_mas_comun = rt_count[0, "room_type"]
contar_mas_comun = rt_count[0, "count"]

print(f"Room type más común: {rt_mas_comun} ({contar_mas_comun} alojamientos)")
```

```python
top10_alcaldias = (
    df.filter(
        pl.col("neighbourhood_cleansed").is_not_null() &
        (pl.col("neighbourhood_cleansed").str.strip_chars() != "")
    )
    .with_columns(
        pl.col("neighbourhood_cleansed")
          .str.strip_chars()
          .alias("neighbourhood_cleansed")
    )
    .group_by("neighbourhood_cleansed")
    .agg(pl.len().alias("total_listings"))
    .sort("total_listings", descending=True)
    .head(10)
)

print(top10_alcaldias)
```

```python
top10_hosts = (
    df.group_by("host_name")
      .len()
      .sort("len", descending=True)
      .head(10)
)

print("Top 10 personas por número de alojamientos:")
print(top10_hosts)
```

```python
# Importación de librerías generación de histograma
import seaborn as sns
import matplotlib.pyplot as plt

# Extraer la columna de Polars como array de NumPy
precios = df["price"].to_numpy()

# Histograma con Seaborn usando el array
plt.figure(figsize=(10, 6))
sns.histplot(precios, bins=50, kde=True, color="red")

plt.title("Distribución de Precios", fontsize=16)
plt.xlabel("Precio", fontsize=10)
plt.ylabel("Frecuencia", fontsize=10)
plt.show()
```

```python
room_counts = (
    df.group_by("room_type")
      .agg(pl.len().alias("total_listings"))
      .sort("total_listings", descending=True)
)

# Extraer datos como listas desde Polars 
tipos = room_counts["room_type"].to_list()
cantidades = room_counts["total_listings"].to_list()

# Graficar con Seaborn usando listas
plt.figure(figsize=(8, 6))
sns.barplot(x=tipos, y=cantidades, palette="viridis")

plt.title("Cantidad de Alojamientos por Tipo", fontsize=14)
plt.xlabel("Tipo de Alojamiento", fontsize=10)
plt.ylabel("Cantidad", fontsize=10)
plt.xticks(rotation=45)
plt.show()
```

```python
# Extraer datos como listas 
alcaldias = top10_alcaldias["neighbourhood_cleansed"].to_list()
cantidades = top10_alcaldias["total_listings"].to_list()

# Gráfica de barras horizontal
plt.figure(figsize=(10, 6))
sns.barplot(x=cantidades, y=alcaldias, palette="viridis")

plt.title("Top 10 Alcaldías con Más Alojamientos", fontsize=16)
plt.xlabel("Cantidad de Alojamientos", fontsize=12)
plt.ylabel("Alcaldía", fontsize=12)
plt.show()
```

```python
import polars as pl
import plotly.graph_objects as go
import plotly.io as pio

# Abrir en el navegador
pio.renderers.default = "browser"

def pick(cols, candidates):
    for c in candidates:
        if c in cols:
            return c
    raise ValueError(f"No fue posible encontrar ninguna de las columnas: {candidates}")

try:
    df
except NameError:
    raise RuntimeError("Se esperaba un dataframe llamado 'df' cargado previamente.")

# Detectamos las columnas
lat_col = pick(df.columns, ["latitude", "lat", "Latitude"])
lon_col = pick(df.columns, ["longitude", "lon", "lng", "Longitude"])
col_precio = pick(df.columns, ["precios", "precio", "price", "Price"])

df_geo = (
    df.with_columns(
        pl.col(col_precio).cast(pl.Utf8).str.replace_all(r"[d\.,-]", "").cast(pl.Float64).alias("price_num")).select([lon_col, lat_col, "price_num"]).drop_nulls()
)

longs = df_geo.select(lon_col).to_series().to_list()
lats = df_geo.select(lat_col).to_series().to_list()
precios = df_geo.select("price_num").to_series().to_list()

if not longs:
    raise ValueError("No se encontraron puntos válidos tras la limpieza.")

pmin, pmax = min(precios), max(precios)
if pmin == pmax:
    sizes = [10] * len(precios)
else:
    s_min, s_max = 5, 14
    sizes = [s_min + (s_max - s_min) * ((p - pmin) / (pmax - pmin)) for p in precios]
    
# Centro del mapa
lat_centro = sum(lats) / len(lats)
lon_centro = sum(longs) / len(longs)

fig = go.Figure(
    go.Scattermapbox(
        lat=lats, lon=longs, mode='markers',
        marker=dict(size=sizes, color=precios, colorscale='Viridis', cmin=pmin, cmax=pmax, opacity=0.7, colorbar=dict(title="Precio")
                    ), hovertemplate="Lat: %{lat}<br>Lon: %{lon}<br>Precio: %{marker.color:.2f}<extra></extra>"
    )
)

fig.update_layout(
    mapbox_style='open-street-map',
    mapbox=dict(center=dict(lat=lat_centro, lon=lon_centro), zoom=9),
    title="Distribución geográfica de Precios",
    margin=dict(r=0, t=50, l=0, b=0)
)

try:
    fig.show()
except Exception as e:
    print("No se pudo abrir el navegador. Guardando 'mapa_precios.html'. Detalle:", e)
    fig.write_html("mapa_precios.html", auto_open=True)
```
