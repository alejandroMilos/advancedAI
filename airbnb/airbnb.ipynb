```python
# Importacion de libreria polars y lectura de archivo CSV
import polars as pl
df = pl.read_csv("./listings.csv")

# Mostrar la tabla en pantalla
df
```

```python
df = df.drop_nulls("price")
print(df.select("price"))
```

```python
# Encontrar el room_type más común
rt_count = df["room_type"].value_counts().sort("count", descending=True)

rt_mas_comun = rt_count[0, "room_type"]
contar_mas_comun = rt_count[0, "count"]

print(f"Room type más común: {rt_mas_comun} ({contar_mas_comun} alojamientos)")
```

```python
nh_count_top10 = (
    df.group_by("neighbourhood")
      .len()
      .sort("len", descending=True)
      .head(10)
)

print("Top 10 alcaldías por número de alojamientos:")
print(nh_count_top10)
```

```python
top10_hosts = (
    df.group_by("host_name")
      .len()
      .sort("len", descending=True)
      .head(10)
)

print("Top 10 personas por número de alojamientos:")
print(top10_hosts)
```

```python
# Importación de librerías generación de histograma
import seaborn as sns
import matplotlib.pyplot as plt

def pick_column(cols, candidates):
    for c in candidates:
        if c in cols:
            return c
    return ValueError(f"No se encontró ninguna de las columnas: {candidates}")

price_col = pick_column(df.columns, ["precios", "precio", "price", "Price", "precio_noche"])

# Limpiamos y convertimos a numérico en Polars
df = df.with_columns (
    pl.col(price_col)
      .cast(pl.Utf8)
      .str.replace_all(r"[^\d\.,-]", "") # quitamos $ y espacios
      .str.replace_all(",", "")          # quitamos cualquier separador de miles
      .cast(pl.Float64, strict=False)
      .alias("price_num")
)

# Quitamos valores nulos
df_hist = df.select("price_num").drop_nulls()

plt.figure(figsize=(8, 5))
sns.histplot(data=df_hist, x="price_num", bins=30)
plt.title("Distribución de Precios")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()
    
```

```python
import pandas as pd

# Tipo de alojamiento vs cantidad
room_col = pick_column(df.columns, ["room_type", "tipo_de_alojamiento", "roomType"])

lat_col = pick_column(df.columns, ["latitude", "lat", "Latitude"])
lon_col = pick_column(df.columns, ["longitude", "lon", "lng", "Longitude"])

counts_room = (
    df.group_by(room_col)
      .agg(pl.len().alias("cantidad"))
      .sort("cantidad", descending=True)
)

plt.figure(figsize=(8, 5))
sns.barplot(data=counts_room, x=room_col, y="cantidad")
plt.title("Tipos de alojamiento vs Cantidad")
plt.xlabel("Tipo de alojamiento")
plt.ylabel("Cantidad de listings")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Alcaldía (usamos la mejor disponible implementando 'fallback')
alcaldia_col = pick_column(
    df.columns,
    ["alcaldia", "alcaldía", "borough", "neighbourhood_group_cleansed",
     "neighbourhood_group", "neighborhood_group", "neighbourhood_cleansed",
     "neighbourhood"]
)

# Barras horizontales - Top 10 alcaldías por cantidad
top10_alcaldias = (
    df.group_by(alcaldia_col)
      .agg(pl.len().alias("cantidad"))
      .sort("cantidad", descending=True)
      .head(10)
      .sort("cantidad")
)

plt.figure(figsize=(8, 6))
sns.barplot(data=top10_alcaldias, x="cantidad", y=alcaldia_col, orient='h')
plt.title(f"Top 10 alcaldías por cantidad de alojamientos (columna: {alcaldia_col})")
plt.xlabel("Cantidad de listings")
plt.ylabel("Alcaldia")
plt.tight_layout()
plt.show()
```
